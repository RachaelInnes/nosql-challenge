### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Coordiantes for Penang Flavours
query = {"BusinessName":"Penang Flavours"}
result = establishments.find_one(query)

latitude_result = result["geocode"]["latitude"]
longitude_result = result["geocode"]["longitude"]

print("Latitude for Penang Flavours:", latitude_result)
print("Longitude for Penang Flavours:", longitude_result)
```

```python
import pymongo
import pprint
import pandas as pd

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client['uk_food']
establishments = db['establishments']

# Remove test data
establishments.delete_many({"BusinessName": {"$in": ["Test Establishment 1", "Test Establishment 2"]}})

# Coordinates for Penang Flavours
query = {"BusinessName": "Penang Flavours"}
result = establishments.find_one(query)

if result:
    # Convert latitude and longitude to floats
    latitude_result = float(result["geocode"]["latitude"])
    longitude_result = float(result["geocode"]["longitude"])

    print("Latitude for Penang Flavours:", latitude_result)
    print("Longitude for Penang Flavours:", longitude_result)

    # Search within 0.01 degree on either side of the latitude and longitude
    degree_search = 0.01

    query = {
        "geocode.latitude": {'$gte': latitude_result - degree_search, '$lte': latitude_result + degree_search},
        "geocode.longitude": {'$gte': longitude_result - degree_search, '$lte': longitude_result + degree_search}
    }
    sort = [("scores.Hygiene", 1)]
    limit = 5

    # Print the query for debugging
    print("Query:", query)

    # Retrieve the results
    results = list(establishments.find(query).sort(sort).limit(limit))
    print("Number of results:", len(results))
    for r in results:
        pprint.pprint(r)

    # Debugging: Print the structure of the first result
    if results:
        print("Structure of the first result:")
        pprint.pprint(results[0])

    # Convert results to a DataFrame
    df = pd.DataFrame(results)

    # Check if 'geocode' and 'scores' columns exist
    if 'geocode' in df.columns and 'scores' in df.columns:
        # Separate geocode into latitude and longitude columns
        df['latitude'] = df['geocode'].apply(lambda x: x['latitude'])
        df['longitude'] = df['geocode'].apply(lambda x: x['longitude'])

        # Separate scores into individual columns
        df['Hygiene'] = df['scores'].apply(lambda x: x['Hygiene'])

        # Drop the original geocode and scores columns
        df = df.drop(columns=['geocode', 'scores'])

        # Display the cleaned DataFrame
        print(df)
    else:
        print("The 'geocode' or 'scores' key is missing in the results.")
else:
    print("No results found for Penang Flavours")

```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {
        '$match': {'scores.Hygiene': 0}
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort': {'count': -1}  # Sorts the matches from highest to lowest
    }
]

# Execute the aggregation pipeline
result = list(db.establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of establishments with a hygiene score of 0:", len(result))

# Print the first 10 results
print("First 10 results:")
for entry in result[:10]:
    print(f"Local Authority: {entry['_id']}, Number of establishments with hygiene score of 0: {entry['count']}")
```

```python
# Convert the result to a Pandas DataFrame
hygiene_zero_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(hygiene_zero_df))

# Display the first 10 rows of the DataFrame
hygiene_zero_df.head(10)
```
